```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('car data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.head(1)
```

```python
date_time = datetime.datetime.now()
df['Age']=date_time.year - df['Year']
```

```python
df.head()
```

```python
df.drop('Year', axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.boxplot(df['Selling_Price'])
```

```python
sorted(df['Selling_Price'],reverse=True)
```

```python
df = df[~(df['Selling_Price']>=33.0) & (df['Selling_Price']<=35.0)]
```

```python
df.shape
```

```python
df.head(1)
```

```python
df['Fuel_Type'].unique()
```

```python
df['Fuel_Type'] = df['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
df['Fuel_Type'].unique()
```

```python
df['Selling_type'].unique()
```

```python
df['Selling_type'] = df['Selling_type'].map({'Dealer':0,'Individual':1})
```

```python
df['Selling_type'].unique()
```

```python
df['Transmission'].unique()
```

```python
df['Transmission'] =df['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
df['Transmission'].unique()
```

```python
df.head()
```

```python
X = df.drop(['Car_Name','Selling_Price'],axis=1)
y = df['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
xgb = GradientBoostingRegressor()
xgb.fit(X_train,y_train)
```

```python
xg = XGBRegressor()
xg.fit(X_train,y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = xgb.predict(X_test)
y_pred4 = xg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
final_data = pd.DataFrame({'Models':['LR','RF','GBR','XG'],
             "R2_SCORE":[score1,score2,score3,score4]})
```

```python
final_data
```

```python
sns.barplot(x=final_data['Models'],y=final_data['R2_SCORE'])
```

```python
xgb = GradientBoostingRegressor()
xgb_final = xgb.fit(X,y)
```

```python
import joblib
```

```python
joblib.dump(xgb_final,'car_price_predictor')
```

```python
model = joblib.load('car_price_predictor')
```
